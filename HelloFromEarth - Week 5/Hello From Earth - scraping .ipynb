```python
import pandas as pd
from bs4 import BeautifulSoup 
import requests
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
numbers = range(1,500)
messages = []

for number in numbers:
    URL = f"https://web.archive.org/web/20130326070614/http://www.hellofromearth.net/gliese581d/messages/{number}/index.html"
    try: 
        response = requests.get(URL) 
        soup = BeautifulSoup(response.content)
        tweets = soup.find_all('div', id='tweet')
        try:
            for tweet in tweets:
                message = {}
                message['Message'] = (tweet.find_all('div')[0].text.strip())
                message['Name'] = (tweet.find_all('div')[1].text.strip())
                message['Location'] = (tweet.find_all('div')[2].text.strip())
                messages.append(message)
        except:
            pass
        print("Scraping", URL)
    except:
        print(URL, "broken")

```

```python
df = pd.DataFrame(messages)
```

```python
df.shape
```

```python
df['City'] = df.Location.str.extract(r'(.*),.*')
```

```python
df['Country']=df.Location.str.extract(r'.*,(.*)')
```
