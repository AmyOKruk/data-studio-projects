```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
import matplotlib.style as style
import matplotlib
from pylab import *
```

```python
df = pd.read_csv("politifacts_score_freq.csv")
```

```python
df.head()
```

```python
# df.transpose()

graph_df = df.pivot(index='Year', columns='Score', values='Percent')
graph_df
```

```python
t = graph_df.index
s = graph_df['False']
s2 = graph_df['Half-True']
s3 = graph_df['Mostly False']
s4 = graph_df['Mostly True']
s5 = graph_df['Pants on Fire!']
s6 = graph_df['True']

ax = plot(t, s, linewidth='3', label='False'),
plot(t, s2, linewidth='3', alpha=0.2),
plot(t, s3, linewidth='3', alpha=0.2),
plot(t, s4, linewidth='3', alpha=0.2),
plot(t, s5, linewidth='3', label='Pants on Fire!'),
plot(t, s6, linewidth='3', label='True')
plt.legend()
```
