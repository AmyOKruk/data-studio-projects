# Scraping PolitiFact's fact checks

This scraper will scrape PolitiFact's fact checks since it started in 2007. 
The organization evaluates statements using a "truth-o-meter" scale, 
ranging from true to pants on fire.

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from time import sleep
```

```python
pages = []
i = 1
while(True):
    try:
        URL = f"https://www.politifact.com/truth-o-meter/statements/?page={i}" 
        print("Scraping", URL)
        response = requests.get(URL) 
        soup = BeautifulSoup(response.content) 
        rows = soup.find_all('div', class_='scoretable__item')
        for row in rows:
            row_dict = {}
            row_dict['Score'] = (row.find('div', class_="meter").a.img.get('alt'))
            row_dict['Date'] = (row.find('span', class_="article__meta").text)
            row_dict['In reality'] = (row.find('div', class_="meter").find('p', class_="quote").text)
            row_dict['Source'] = (row.find('div', class_="statement__source").text.strip())
            row_dict['Statement'] = (row.find('p', class_="statement__text").text.strip())
            row_dict['Link'] = (row.find('p', class_="statement__text").a.get('href'))
            row_dict['PolitiFact source'] = (row.find('p', class_="statement__edition").a.text.strip())
            pages.append(row_dict)
        i = i+1
        sleep(0.25)
    except:
        break
```

```python
df = pd.DataFrame(pages)
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv("politifacts_data.csv")
```

# Preparing the data for graphing/analysis

Some years, Politifact fact checked more statements than others. Therefore, I want to look at the year-to-year percentage of "pants on fire" statements, not the total number. I'm working with value counts to create a new dataframe that shows how frequent each meter score was per year.


```python
df['Year'] = df.Date.str.extract(r'(\d{4})')
```

```python
vc = df.Year.value_counts()
```

```python
df2 = pd.DataFrame(df.groupby(by='Year').Score.value_counts())
```

```python
df2 = df2.rename(columns={'Score':'Count'})
```

```python
df2 = df2.reset_index()
```

```python
df2 = df2.set_index('Year')
```

```python
percents = df2.Count / vc
```

```python
df2['Percent'] = percents
```

```python
df2.Percent = df2.Percent*100
```

```python
df2.head()
```

Removing statements scored as Full Flop, Half Flip, No Flip

```python
df2 = df2[(df2.Score != 'Full Flop') & (df2.Score != 'Half Flip') & (df2.Score != 'No Flip')]
```

```python
df2 = df2.reset_index()
```

```python
df2.to_csv("politifacts_score_freq.csv", index=False)
```
