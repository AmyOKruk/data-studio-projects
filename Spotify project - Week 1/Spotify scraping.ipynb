```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
```

```python
# Scraping Spotify Top 200 songs per country for week of 06/21/2019 to 06/28/2019
# Andorra (ad) and Monaco (mc) do not return results 
```

```python
countries = ['us','gb','ad','ar','at','au','be','bg','bo','br','ca','ch','cl','co','cr','cy','cz','de','dk','do','ec','ee','es','fi',
            'fr','gr','gt','hk','hn','hu','id','ie','il','in','is','it','jp','lt','lu','lv','mc','mt','mx','my','ni','nl','no','nz','pa','pe','ph',
            'pl','pt','py','ro','se','sg','sk','sv','th','tr','tw','uy','vn','za']

spotify_data = []

for country in countries:
    try:
        URL = f"https://spotifycharts.com/regional/{country}/weekly/2019-06-21--2019-06-28"
        response = requests.get(URL) 
        soup = BeautifulSoup(response.content) 
        table = soup.find('tbody').find_all('tr')
        country_dict = {}
        for row in table:
            rank = row.find_all('td')[1].string
            name = row.find_all('td')[3].strong.text + " "+ row.find_all('td')[3].span.text
            country_dict[rank] = name
        spotify_data.append(country_dict)
    except:
        print("Error for", country)

df = pd.DataFrame(spotify_data)

df.head()
```

```python
df = df.rename(index={0:'us',1:'gb',2:'ar',3:'at',4:'au',5:'be',6:'bg',7:'bo',8:'br',9:'ca',10:'ch',11:'cl',12:'co',13:'cr',14:'cy',15:'cz',16:'de',17:'dk',18:'do',19:'ec',20:'ee',21:'es',22:'fi',
            23:'fr',24:'gr',25:'gt',26:'hk',27:'hn',28:'hu',29:'id',30:'ie',31:'il',32:'in',33:'is',34:'it',35:'jp',36:'lt',37:'lu',38:'lv',39:'mt',40:'mx',41:'my',42:'ni',43:'nl',44:'no',45:'nz',46:'pa',47:'pe',48:'ph',
            49:'pl',50:'pt',51:'py',52:'ro',53:'se',54:'sg',55:'sk',56:'sv',57:'th',58:'tr',59:'tw',60:'uy',61:'vn',62:'za'})
```
